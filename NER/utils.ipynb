```python
# plot for 1 layer 
relu = pd.read_csv('./results/1layer-1d-relu.csv')
norm = pd.read_csv('./results/1layer-1d.csv')
index = relu.index
epoch = index+1
```

```python
plt.figure(figsize=(10,6))
plt.plot(epoch, norm['test_score'])
plt.plot(epoch, relu['test_score'])
plt.legend(['Without ReLU', 'With ReLU'], loc = 'best')
plt.title('Test Accuracy with/without ReLU') 
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.show()
```
